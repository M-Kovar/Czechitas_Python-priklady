## 02a: Základní úpravy a prvotní analýza - cvičení

Našli jsme si ideální dataset pro projekt, který má za cíl analyzovat populaci světových hlavních měst. CSV ale obsahuje sloupce, které nás nezajímají, a chybějící hodnoty, proto data před použitím upravíme. Pomocí pandas proveď následující zpracování přiloženého datasetu [worldcities.csv](worldcities.csv) (zdroj: https://simplemaps.com/data/world-cities):

*Důležité je se před využitím jakéhokoliv datasetu dobře seznámit s významem sloupců, zdrojem dat, atd., což nabízí odkaz výše. Pro úsporu času toto ale můžeš přeskočit, detaily pro cvičení nemusíš znát.*

- Načti dataset jako DataFrame a podívej se, jak data vypadají, např. metodou `sample()` pro náhodný výběr řádků.
- Redukuj dataset pouze na sloupce "city", "country", "capital" a "population".
- Odstraň všechny řádky obsahující chybějící hodnoty (NaN). Ověř, kolik zbylo záznamů z původního počtu.
- Zobraz počet měst podle jednotlivých kategorií v sloupci "capital". Zajímají nás hlavně "primary" capitals (státní hlavní města); "admin" a "minor" značí hlavní města nižších správních celků. Dává počet "primary" zhruba smysl?
- Proveď agregaci dat metodou `groupby()` podle typu hlavního města (sloupec "capital") a pomocí metody `median()` zobraz střední hodnotu populace v každé kategorii. Když porovnáš kategorie, jsou hodnoty rozložené podle tvého očekávání?
- Chceme zachovat pouze státní hlavní města. Vyber tedy pouze řádky, kde sloupec "capital" obsahuje hodnotu "primary".
- Ulož upravený dataset do souboru `worldcities_reduced.csv` bez indexu.

```python
# Čištění dat

import pandas as pd

data = pd.read_csv("worldcities.csv")
data
```

```python
# Náhodný výběr záznamů
data.sample(10)
```

```python
# Redukce sloupců
data = data[["city", "country", "capital", "population"]]
# data = data.drop(["city_ascii", "lat", "lng", "iso2", "iso3", "admin_name", "id"], axis=1)
data
```

```python
# Odstranění chybějících hodnot
data = data.dropna()
# data = data.dropna(subset=["population", "capital"])
data
```

```python
# Zobrazení počtů kategorií měst
count_by_category = data["capital"].value_counts()
count_by_category
```

```python
# Agregace podle typu města
median_by_category = data.groupby("capital")["population"].median()
median_by_category
```

```python
# Výběr pouze primárních hlavních měst
data_primary = data[data["capital"] == "primary"]
data_primary
```

```python
# Uložení do csv (bez indexu)
data_primary.to_csv("worldcities_reduced.csv", index=False)
```
